```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
#read data
data=pd.read_csv("/content/Student_performance_data _.csv")
```

```python
data.info()
```

```python

data.head()
```

```python
data=data.drop("StudentID",axis=1)
```

```python
sns.barplot(x="ParentalSupport",y="Absences",data=data)
plt.title("ParentalSupport vs Absences")
plt.xlabel("ParentalSupport")
plt.ylabel("Absences")

plt.show()
```

```python
#data visualization

plt.figure(figsize=(10,5))
sns.heatmap(data.corr(),annot=True)
```

```python
sns.pairplot(data)
```

```python
#model
model=LogisticRegression()
```

```python
features=data.drop("GradeClass",axis=1)
label=data["GradeClass"]
```

```python
feature_train,feature_test,label_train,label_test=train_test_split(features,label,test_size=0.5,random_state=20)
```

```python
model.fit(feature_train,label_train)
```

```python

prediction=model.predict(feature_test)
prediction=pd.DataFrame(prediction)
prediction = prediction.round(decimals=0).astype(float)
prediction.rename(columns={0:"prediction"},inplace=True)
prediction.head()
```

```python
label_test=label_test.reset_index(drop=True)
label_test=pd.DataFrame(label_test)
label_test.rename(columns={"GradeClass":"label_test"},inplace=True)
label_test.head()
```

```python
new_data = pd.concat([label_test, prediction], axis = 1, join = 'inner')
new_data.head()
```
